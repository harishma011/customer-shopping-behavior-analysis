## 2.9 Derived Columns

### 2.9.1 Derived columns - Age Group

```python
df['Age Group'] = pd.cut(
    df['Age'],
    bins=[18, 25, 35, 45, 55, 70],
    labels=['18-25', '26-35', '36-45', '46-55', '55+']
)
```

### Purpose

To categorize customers into different age groups for easier demographic analysis.

### Observation

A new column named Age Group was created from the Age column. Customers were grouped into five age categories.

### Interpretation

Age grouping helps identify which age segments contribute most to sales and customer engagement.

### 2.9.2 Derived Columns - Customer Type

```python
df['Customer Type'] = np.where(
    df['Previous Purchases'] > 25,
    'Returning Customer',
    'New Customer'
)
```

### Purpose

To classify customers based on their purchase history.

### Observation

Customers with more than 25 previous purchases were categorized as Returning Customers, while the remaining customers were categorized as New Customers.

### Interpretation

This classification helps analyze customer loyalty and compare purchasing behavior between new and returning customers.

### 2.9.3 Derived Columns - Spending Category

```python
df['Spending Category'] = pd.cut(
    df['Purchase Amount (USD)'],
    bins=[0, 40, 70, 100],
    labels=['Low', 'Medium', 'High']
)
```

### Purpose

To segment customers based on their purchase amount.

### Observation

Customers were classified into Low, Medium, and High spending categories according to their purchase amount.

### Interpretation

Spending categories help identify high-value customers and support targeted marketing strategies.

### 2.9.4 Verify Derived Columns
